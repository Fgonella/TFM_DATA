# Clasificación Estabulado vs Pastoreo — Provincia de Huesca
## Metodología del análisis de stock ganadero para planta de biometano
 
**Fecha de referencia:** 1 de noviembre de 2024  
**Fuente de datos:** MAPA — Encuesta Anual de Ganado Ovino y Caprino; Encuesta Semestral de Ganado Bovino y Porcino  
**Base legal:** Reglamento (CE) n.º 1165/2008 · Reglamento de Ejecución (UE) 2023/2745  
**Registro administrativo:** REGA (Registro General de Explotaciones Ganaderas)

## 1. Objetivo
 
Cuantificar el stock ganadero adulto en producción en la provincia de Huesca, distinguiendo entre animales estabulados y en pastoreo, con el fin de estimar el potencial de feedstock (purín/estiércol) para una planta de biometano.

## 2. Decisiones metodológicas previas
 
### 2.1 Fecha homogénea para todas las especies
 
Las encuestas ganaderas del MAPA tienen frecuencias de publicación distintas según especie:
 
| Especie | Frecuencia | Última fecha disponible |
|---|---|---|
| Bovino | Semestral (mayo y noviembre) | Mayo 2025 |
| Porcino | Semestral (mayo y noviembre) | Mayo 2025 |
| Ovino | Anual (noviembre) | Noviembre 2024 |
| Caprino | Anual (noviembre) | Noviembre 2024 |
 
Se optó por **noviembre de 2024** como fecha de referencia común para todas las especies. Utilizar fechas distintas por especie comprometería la comparabilidad del análisis y su validez metodológica.

### 2.2 Exclusión de animales jóvenes y de cebo
 
Para el objetivo del estudio — estimar un feedstock estable y predecible para una planta de biometano — se excluyen los animales jóvenes y de engorde por las siguientes razones:
 
- **Alta rotación**: El censo de animales de cebo varía significativamente entre periodos, generando volatilidad en la estimación de purín disponible.
- **Baja predictibilidad**: La composición y volumen de purín de animales jóvenes es muy variable e inferior a la de adultos en producción.
- **Inestabilidad contractual**: Una planta de biometano requiere compromisos de suministro a largo plazo; los animales de cebo no garantizan continuidad.


## 3. Clasificación por especie
 
La clasificación sigue las definiciones oficiales del **Informe Metodológico Estandarizado (IME)** del MAPA, sección 3.4, que a su vez implementa las categorías del Reglamento (CE) n.º 1165/2008.

### 3.1 Bovino
 
Según el IME Bovino §3.4, las categorías clave son:
 
- **Vacas lecheras**: *"Vacas destinadas exclusiva o principalmente a la producción de leche"* → **Estabulado**
- **Otras vacas**: *"Vacas distintas de las lecheras, incluidas las vacas de labor"* → **Pastoreo**
- **Otras novillas**: *"Novillas criadas para la reproducción"* → **Pastoreo**
- **Novillas de abasto**: *"Novillas criadas para la producción de carne"* → **Excluido**
| Variable | Descripción | Clasificación | Justificación |
|---|---|---|---|
| `11ANIMAYOR24VL` | Vacas lecheras | Estabulado | IME Bovino §3.4 explícito |
| `12ANIMAYOR24VR` | Otras vacas (nodrizas) | Pastoreo | IME Bovino §3.4 explícito |
| `07ANI12A24HR` | Novillas reproductoras 12-24m | Pastoreo | IME Bovino §3.4 "Otras novillas" |
| `10ANIMAYOR24NR` | Novillas reproductoras >24m | Pastoreo | IME Bovino §3.4 "Otras novillas" |
| `08ANIMAYOR24M` | Machos adultos >24m (toros/bueyes) | Pastoreo | Inferencia territorial: bovino pirenaico extensivo en Huesca |
| `02ANI0A12DS` | Terneros/jóvenes de abasto <12m | Excluido | Alta rotación, feedstock volátil |
| `03ANI0A12OM` | Machos <12m | Excluido | Alta rotación, feedstock volátil |
| `04ANI0A12OH` | Hembras <12m | Excluido | Alta rotación, feedstock volátil |
| `05ANI12A24M` | Machos de engorde 12-24m | Excluido | Cebo, no permanentes |
| `06ANI12A24HS` | Novillas de abasto 12-24m | Excluido | IME Bovino §3.4 "Novillas de abasto" |
| `09ANIMAYOR24NS` | Novillas de sacrificio >24m | Excluido | Destino sacrificio, no permanentes |
 
> **Nota sobre `08ANIMAYOR24M`**: El IME define toros y bueyes sin asignarles sistema productivo explícito. Se clasifican como pastoreo por coherencia con el modelo extensivo predominante del bovino en la provincia de Huesca (razas pirenaicas, Parda de Montaña), donde los machos adultos acompañan al rebaño en aprovechamiento de pastos de montaña.
 
---
 
### 3.2 Ovino
 
Según el IME Ovino/Caprino §3.4, las categorías oficiales son:
 
- **Ovejas lecheras**: *"Ovejas criadas principal o exclusivamente para la producción de leche"* → **Estabulado**
- **Otras ovejas**: *"Ovejas distintas a las ovejas lecheras"* → **Pastoreo**
- **Corderos**: *"Animales jóvenes hasta 12 meses"* → **Excluido**
- **Sementales**: *"Machos no castrados"* → **Pastoreo**

El sufijo **O/NO** en los códigos de variable indica orientación productiva:
- **O** = "Ovejas" → sistema extensivo (cárnico) → **Pastoreo**
- **NO** = "No Ovejas" → sistema intensivo (lácteo) → **Estabulado**

| Variable | Descripción | Clasificación | Justificación |
|---|---|---|---|
| `09HPNO` | Hembras paridas, sistema lácteo | Estabulado | IME §3.4 "Ovejas lecheras" · sufijo NO |
| `07HNPC1VNO` | Hembras lácteas, 1ª cubrición | Estabulado | IME §3.4 "Ovejas lecheras" · sufijo NO |
| `08HPO` | Hembras paridas, sistema cárnico | Pastoreo | IME §3.4 "Otras ovejas" · sufijo O |
| `06HNPC1VO` | Hembras cárnicas, 1ª cubrición | Pastoreo | IME §3.4 "Otras ovejas" · sufijo O |
| `03Sementales` | Machos reproductores | Pastoreo | IME §3.4 "Sementales" |
| `02Corderos` | Jóvenes <12m | Excluido | IME §3.4 "Corderos" · alta rotación |
| `05HNPNC` | Hembras adultas sin cubrir | **Prorrateo provincial** | Ver nota |
 
> **Nota sobre `05HNPNC`**: Las Hembras No Paridas No Cubiertas (39.750 cabezas) son animales adultos en producción cuya orientación —lechera o cárnica— no puede determinarse directamente de los datos publicados, ya que el MAPA no desglosa esta variable con sufijo O/NO ni publica microdatos (IME §10.4). En lugar de excluirlas, se distribuyen mediante **prorrateo proporcional** al ratio estabulado/pastoreo observado en las hembras ovinas clasificadas de la provincia de Huesca (`09HPNO + 07HNPC1VNO` vs `08HPO + 06HNPC1VO + 03Sementales`). Este enfoque asume que las `05HNPNC` comparten la misma distribución por sistema de manejo que la cabaña ovina provincial, coherente con el principio de representatividad territorial ya aplicado en otras variables (véase §3.1, nota `08ANIMAYOR24M`). El ratio se calcula dinámicamente en el código a partir de los datos reales.
 
---
 
### 3.3 Caprino
 
Se aplica la misma lógica que en ovino. El sufijo O/NO identifica orientación productiva donde aparece.
 
| Variable | Descripción | Clasificación | Justificación |
|---|---|---|---|
| `09HPNO` | Hembras paridas, sistema lácteo | Estabulado | IME §3.4 "Cabras lecheras" · sufijo NO |
| `08HPO` | Hembras paridas, sistema extensivo | Pastoreo | IME §3.4 "Otras cabras" · sufijo O |
| `03Sementales` | Machos reproductores | Pastoreo | IME §3.4 "Sementales" |
| `05HNPNC` | Hembras adultas sin cubrir | Pastoreo | Volumen marginal (1,214 cab., 6.3% del total caprino) |
| `06HNPC1V` | Hembras, 1ª cubrición | Pastoreo | Ver nota |
| `02Chivos` | Jóvenes destino sacrificio | Excluido | IME §3.4 "Chivos" · destino sacrificio |
 
> **Nota sobre `06HNPC1V`**: A diferencia del ovino, el dataset no desglosa esta variable con sufijo O/NO para caprino, impidiendo distinguir sistema lechero de extensivo. Se clasifica como pastoreo dado el perfil predominantemente extensivo del caprino en Huesca y su volumen marginal (1,419 cabezas, 7.4% del total caprino). El impacto sobre el análisis global es inferior al 0.03%.
 
---
 
### 3.4 Porcino
 
El porcino se clasifica íntegramente como estabulado. El sistema de producción porcina intensiva no contempla pastoreo en ninguna de sus categorías productivas. Se utiliza `01Tanimales` (total censal) como variable agregada para evitar la doble contabilización que producirían las variables NI (14-26) e IB (27-39), que desglosan el mismo total por tipo de integración.

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
from tabulate import tabulate
from pathlib import Path

RAW_DIR = Path("../../data/raw/01_ganado_porcino")
MAP_DIR = Path("../../data/map/01_ganado_porcino")
RAW_DIR.mkdir(parents=True, exist_ok=True)
MAP_DIR.mkdir(parents=True, exist_ok=True)

# --- Carga y limpieza ---
df = pd.read_csv(RAW_DIR / 'EFECTIVOS_GANADO - data.csv')
df['VALOR'] = pd.to_numeric(df['VALOR'].astype(str).str.replace(',', '.'), errors='coerce').fillna(0)

# --- Filtro: Porcino total, mayo 2025 ---
porcino = df[
    (df['AÑO'] == 2025) &
    (df['MES'] == 5) &
    (df['ESPECIE'] == 'Porcino') &
    (df['VARIABLE'] == '01Tanimales')
].copy()

total_españa = porcino[porcino['COMUNIDAD'] == 99]['VALOR'].sum()

# --- Limpiar nombres (quitar código numérico del principio) ---
def limpiar_nombre(nombre):
    partes = nombre.split(' ', 1)
    return partes[1] if len(partes) > 1 else nombre

# --- CCAA ---
ccaa = (
    porcino[(porcino['PROVINCIA'] == 99) & (porcino['COMUNIDAD'] != 99)]
    [['NOMBRE DE LA COMUNIDAD-PROVINCIA', 'VALOR']]
    .sort_values('VALOR', ascending=False)
    .reset_index(drop=True)
)
ccaa['COMUNIDAD'] = ccaa['NOMBRE DE LA COMUNIDAD-PROVINCIA'].apply(limpiar_nombre)
ccaa['CABEZAS'] = ccaa['VALOR'].apply(lambda x: f"{x:,.0f}")
ccaa['% ESPAÑA'] = (ccaa['VALOR'] / total_españa * 100).apply(lambda x: f"{x:.1f}%")

# --- Provincias ---
provs = (
    porcino[porcino['PROVINCIA'] != 99]
    [['NOMBRE DE LA COMUNIDAD-PROVINCIA', 'VALOR']]
    .sort_values('VALOR', ascending=False)
    .reset_index(drop=True)
)
provs['PROVINCIA'] = provs['NOMBRE DE LA COMUNIDAD-PROVINCIA'].apply(limpiar_nombre)
provs['CABEZAS'] = provs['VALOR'].apply(lambda x: f"{x:,.0f}")
provs['% ESPAÑA'] = (provs['VALOR'] / total_españa * 100).apply(lambda x: f"{x:.1f}%")

# --- Tablas en consola ---
print(f"\n{'='*55}")
print(f"  PORCINO ESPAÑA — MAYO 2025  |  Total: {total_españa:,.0f} cab.")
print(f"{'='*55}\n")

print("RANKING POR COMUNIDAD AUTÓNOMA")
print(tabulate(
    ccaa[['COMUNIDAD', 'CABEZAS', '% ESPAÑA']],
    headers=['Comunidad Autónoma', 'Cabezas', '% España'],
    tablefmt='rounded_outline',
    showindex=True
))

print("\nRANKING POR PROVINCIA")
print(tabulate(
    provs[['PROVINCIA', 'CABEZAS', '% ESPAÑA']],
    headers=['Provincia', 'Cabezas', '% España'],
    tablefmt='rounded_outline',
    showindex=True
))

# --- Gráficos ---
fig, axes = plt.subplots(1, 2, figsize=(18, 7))
fig.suptitle('Censo Porcino España — Mayo 2025', fontsize=15, fontweight='bold', y=1.01)

# Colores
color_principal = '#e07b39'
color_resto = '#f5c6a0'

def plot_ranking(ax, data, col_nombre, titulo, top_n=None):
    d = data if top_n is None else data.head(top_n)
    colores = [color_principal if i == 0 else color_resto for i in range(len(d))]
    bars = ax.barh(d[col_nombre][::-1], d['VALOR'][::-1] / 1e6, color=colores[::-1])
    ax.set_title(titulo, fontweight='bold', fontsize=12, pad=10)
    ax.set_xlabel('Millones de cabezas', fontsize=10)
    ax.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{x:.1f}M'))
    ax.spines[['top', 'right']].set_visible(False)
    # Etiquetas en las barras
    for bar, (_, row) in zip(bars[::-1], d.iterrows()):
        pct = row['VALOR'] / total_españa * 100
        ax.text(bar.get_width() + 0.02, bar.get_y() + bar.get_height()/2,
                f"{pct:.1f}%", va='center', fontsize=8.5, color='#444')

plot_ranking(axes[0], ccaa, 'COMUNIDAD', 'Por Comunidad Autónoma')
plot_ranking(axes[1], provs, 'PROVINCIA', 'Top 15 Provincias', top_n=15)

plt.tight_layout()
plt.savefig(MAP_DIR / 'porcino_ranking_2025.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
from tabulate import tabulate

# --- Filtro: Huesca, datos más recientes por especie ---
huesca = df[df['NOMBRE DE LA COMUNIDAD-PROVINCIA'].str.contains('Huesca', case=False, na=False)]

# Solo totales de cada especie (variable 01Tanimales) y el dato más reciente
huesca_total = (
    huesca[(huesca['VARIABLE'] == '01Tanimales') &
    (huesca['AÑO'] == 2024) &
    (huesca['MES'] == 11)
].copy())

huesca_total = huesca_total[huesca_total['VALOR'] > 0].sort_values('VALOR', ascending=False)

# --- Significancia ---
total_cabezas = huesca_total['VALOR'].sum()
huesca_total['Significancia (%)'] = (huesca_total['VALOR'] / total_cabezas * 100).round(2)

# --- Tabla ---
tabla = huesca_total[['ESPECIE', 'VALOR', 'AÑO', 'MES', 'Significancia (%)']].copy()
tabla.columns = ['Especie', 'Cabezas', 'Año', 'Mes', 'Significancia (%)']
tabla['Cabezas'] = tabla['Cabezas'].astype(int)

print(f"\n{'='*55}")
print(f"  INVENTARIO GANADERO — HUESCA (Últimos datos disponibles)")
print(f"  Total: {int(total_cabezas):,} cabezas")
print(f"{'='*55}")
print(tabulate(tabla.reset_index(drop=True), headers='keys', tablefmt='rounded_outline', floatfmt=".2f"))
# --- Gráfico de barras horizontal ---
fig, ax = plt.subplots(figsize=(10, 5))
colors = plt.cm.tab10.colors

bars = ax.barh(huesca_total['ESPECIE'], huesca_total['VALOR'], color=colors[:len(huesca_total)])

ax.set_xlabel('Número de cabezas', fontsize=11)
ax.set_title('Activos ganaderos por especie — Provincia de Huesca', fontsize=13, fontweight='bold')
ax.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{int(x):,}'))
ax.invert_yaxis()

# Etiquetas con valor y porcentaje
for bar, (_, row) in zip(bars, huesca_total.iterrows()):
    ax.text(bar.get_width() + total_cabezas * 0.005, bar.get_y() + bar.get_height() / 2,
            f"{int(row['VALOR']):,}  ({row['Significancia (%)']:.1f}%)",
            va='center', fontsize=9)

ax.set_xlim(0, huesca_total['VALOR'].max() * 1.25)
plt.tight_layout()
plt.savefig(MAP_DIR / 'huesca_ganadero_especies.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# ============================================================
# ESTABULADO vs PASTOREO — HUESCA (11/2024)
# Fuente: MAPA — Encuesta Anual Ovino/Caprino & Semestral Bovino
# Clasificación basada en IME MAPA §3.4 (Reglamento CE 1165/2008)
# ============================================================

# --- Filtro: Huesca, noviembre 2024 (fecha homogénea todas las especies) ---
_huesca = df[
    df['NOMBRE DE LA COMUNIDAD-PROVINCIA'].str.contains('Huesca', case=False, na=False)
].copy()
_huesca = _huesca[(_huesca['AÑO'] == 2024) & (_huesca['MES'] == 11)]

# --- Clasificación definitiva con justificación metodológica ---
vars_clasificadas = {
    'Bovino': {
        'estabulado': ['11ANIMAYOR24VL'],
        'pastoreo':   ['12ANIMAYOR24VR', '07ANI12A24HR',
                       '08ANIMAYOR24M',   # Toros/bueyes → pastoreo por contexto territorial pirenaico
                       '10ANIMAYOR24NR'],
        'excluido':   ['02ANI0A12DS', '03ANI0A12OM', '04ANI0A12OH',
                       '05ANI12A24M', '06ANI12A24HS', '09ANIMAYOR24NS']
    },
    'Ovino': {
        'estabulado': ['09HPNO', '07HNPC1VNO'],
        'pastoreo':   ['03Sementales', '06HNPC1VO', '08HPO'],
        'prorrateo':  ['05HNPNC'],   # Prorrateo: ratio provincial observado en hembras clasificadas
        'excluido':   ['02Corderos']
    },
    'Caprino': {
        'estabulado': ['09HPNO'],
        'pastoreo':   ['03Sementales', '05HNPNC',
                       '06HNPC1V',   # Pastoreo por volumen marginal (7.4%) y perfil extensivo Huesca
                       '08HPO'],
        'excluido':   ['02Chivos']
    },
    'Porcino': {
        'estabulado': ['01Tanimales'],
        'pastoreo':   [],
        'excluido':   []
    }
}

# --- Verificación de integridad ANTES de calcular ---
print("AUDITORÍA PRE-EJECUCIÓN")
print("="*55)
ok = True
for esp, clasif in vars_clasificadas.items():
    sub = _huesca[_huesca['ESPECIE'] == esp]
    total_real = sub[sub['VARIABLE'] == '01Tanimales']['VALOR'].sum()
    todas = clasif['estabulado'] + clasif['pastoreo'] + clasif.get('prorrateo', []) + clasif['excluido']
    total_nuestro = sub[sub['VARIABLE'].isin(todas)]['VALOR'].sum()
    diff = total_real - total_nuestro
    estado = '✅' if abs(diff) < 1 else f'⚠️  DIFF={diff:,.0f}'
    print(f"  {estado}  {esp}: 01T={total_real:,.0f} | clasificado={total_nuestro:,.0f}")
    if abs(diff) >= 1:
        ok = False
print("="*55)
print(f"  {'✅ Auditoría OK — procedemos' if ok else '⚠️  Revisar antes de continuar'}")
print()

# --- Cálculo solo si auditoría OK ---
if ok:
    resumen = []
    for esp, clasif in vars_clasificadas.items():
        sub  = _huesca[_huesca['ESPECIE'] == esp]
        est  = sub[sub['VARIABLE'].isin(clasif['estabulado'])]['VALOR'].sum()
        past = sub[sub['VARIABLE'].isin(clasif['pastoreo'])]['VALOR'].sum()

        # Prorrateo para variables sin sufijo O/NO (aplica a Ovino 05HNPNC)
        if 'prorrateo' in clasif and clasif['prorrateo']:
            n_prorrateo = sub[sub['VARIABLE'].isin(clasif['prorrateo'])]['VALOR'].sum()
            base = est + past
            if base > 0 and n_prorrateo > 0:
                ratio_est  = est  / base
                ratio_past = past / base
                est  += n_prorrateo * ratio_est
                past += n_prorrateo * ratio_past
                print(f"  [{esp}] 05HNPNC prorrateo: {n_prorrateo:,.0f} cab. → "                      f"+{n_prorrateo*ratio_est:,.0f} estab. / +{n_prorrateo*ratio_past:,.0f} past. "                      f"(ratio {ratio_est*100:.1f}% / {ratio_past*100:.1f}%)")

        resumen.append({'Especie': esp, 'Estabulado': est, 'Pastoreo': past})

    resultado = pd.DataFrame(resumen)
    resultado['Total']        = resultado['Estabulado'] + resultado['Pastoreo']
    resultado['% Estabulado'] = (resultado['Estabulado'] / resultado['Total'].where(resultado['Total'] > 0) * 100).fillna(0).round(1)
    resultado['% Pastoreo']   = (resultado['Pastoreo']   / resultado['Total'].where(resultado['Total'] > 0) * 100).fillna(0).round(1)

    total_est  = resultado['Estabulado'].sum()
    total_past = resultado['Pastoreo'].sum()
    total_all  = total_est + total_past

    print(f"\n{'='*65}")
    print(f"  ESTABULADO vs PASTOREO — HUESCA (11/2024)")
    print(f"  Fuente: MAPA REGA · Clasificación: IME §3.4 (CE 1165/2008)")
    print(f"  Excluidos: jóvenes/cebo")
    print(f"  Estabulado: {int(total_est):,}  |  Pastoreo: {int(total_past):,}")
    print(f"  % Estabulado global: {total_est/total_all*100:.1f}%")
    print(f"{'='*65}")
    print(tabulate(resultado.reset_index(drop=True),
                   headers='keys', tablefmt='rounded_outline', floatfmt=".1f"))

    # --- Gráfico ---
    fig, ax = plt.subplots(figsize=(10, 4))
    bars_est  = ax.barh(resultado['Especie'], resultado['Estabulado'],
                        color='#2E75B6', label='Estabulado')
    bars_past = ax.barh(resultado['Especie'], resultado['Pastoreo'],
                        left=resultado['Estabulado'], color='#70AD47', label='Pastoreo')

    for i, row in resultado.iterrows():
        if row['Estabulado'] > 0:
            ax.text(row['Estabulado'] / 2, i, f"{row['% Estabulado']:.0f}%",
                    va='center', ha='center', color='white', fontweight='bold', fontsize=9)
        if row['Pastoreo'] > 0:
            ax.text(row['Estabulado'] + row['Pastoreo'] / 2, i, f"{row['% Pastoreo']:.0f}%",
                    va='center', ha='center', color='white', fontweight='bold', fontsize=9)

    ax.set_xlabel('Número de cabezas (adultos en producción)', fontsize=11)
    ax.set_title('Estabulado vs Pastoreo — Huesca (11/2024)\n'
                 'Clasificación MAPA IME §3.4 · Excluidos jóvenes/cebo',
                 fontsize=12, fontweight='bold')
    ax.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{int(x):,}'))
    ax.legend(loc='lower right')
    ax.invert_yaxis()
    plt.tight_layout()
    plt.savefig(MAP_DIR / 'huesca_estabulado_pastoreo_definitivo.png', dpi=150, bbox_inches='tight')
    plt.show()
else:
    print("  Corrige las diferencias antes de continuar.")
```

```python
"""
KDE densidad porcina — Huesca
Fuente: REGA Aragón Open Data (CC BY 4.0)
https://opendata.aragon.es/datos/catalogo/dataset/
        datos-del-registro-de-explotaciones-ganaderas-de-aragon-rega-y-de-nucleos-zoologicos

Dato MAPA nov-2024: 5.099.794 cerdos reales
Capacidad REGA:     5.873.367 plazas autorizadas (~115% — normal)
"""

import pandas as pd
import numpy as np
import geopandas as gpd
from pyproj import Transformer
from scipy.stats import gaussian_kde
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from shapely.geometry import Point
from shapely.ops import unary_union
from pathlib import Path
import rasterio
from rasterio.transform import from_bounds

# ─────────────────────────────────────────────────────────
# CONFIGURACIÓN
# ─────────────────────────────────────────────────────────
RAW_DIR        = Path("../../data/raw/01_ganado_porcino")
MAP_DIR        = Path("../../data/map/01_ganado_porcino")
RAW_DIR.mkdir(parents=True, exist_ok=True)
MAP_DIR.mkdir(parents=True, exist_ok=True)
CSV_PORCINO    = RAW_DIR / "PORCINO.csv"
DELIM_DIR      = Path("../../data/raw/delimitations")
GEOJSON_HUESCA = DELIM_DIR / "Huesca_Delimitacion.geojson"
OUTPUT_TIF     = RAW_DIR / "kde_porcino_huesca.tif"
OUTPUT_PNG     = MAP_DIR / "kde_porcino_huesca.png"
OUTPUT_GPKG    = RAW_DIR / "granjas_porcino_huesca.gpkg"

CRS_UTM = "EPSG:25830"
RES_M   = 500   # resolución del raster KDE en metros

# ─────────────────────────────────────────────────────────
# 1. CARGAR Y LIMPIAR DATOS
# ─────────────────────────────────────────────────────────
print("Cargando datos REGA porcino...")
df = pd.read_csv(CSV_PORCINO)

# Filtrar Huesca
df = df[df["provincia"] == "HUESCA"].copy()
print(f"  Granjas Huesca: {len(df)}")

# Eliminar filas sin coordenadas
df = df.dropna(subset=["longitud_utm", "latitud_utm"])
print(f"  Con coordenadas: {len(df)}")

# Calcular capacidad total por granja (suma de todas las plazas)
cols_cap = ["capacidad_cebo", "capacidad_lechones", "capacidad_recria",
            "capacidad_cerdas", "capacidad_reposicion", "capacidad_verracos"]
df["capacidad_total"] = df[cols_cap].fillna(0).sum(axis=1)

# Eliminar granjas con capacidad 0
df = df[df["capacidad_total"] > 0].copy()
print(f"  Con capacidad > 0: {len(df)}")
print(f"  Capacidad total provincia: {df['capacidad_total'].sum():,.0f} plazas")

# ─────────────────────────────────────────────────────────
# 2. CONVERTIR COORDENADAS UTM → EPSG:25830
# ─────────────────────────────────────────────────────────
print("Convirtiendo coordenadas UTM 30N → EPSG:25830...")

# Los datos están en UTM huso 30 (EPSG:32630 = WGS84 UTM30N)
# EPSG:25830 = ETRS89 UTM30N — prácticamente idéntico para España
transformer = Transformer.from_crs("EPSG:32630", CRS_UTM, always_xy=True)
x_utm, y_utm = transformer.transform(
    df["longitud_utm"].values,
    df["latitud_utm"].values
)
df["x"] = x_utm
df["y"] = y_utm

# ─────────────────────────────────────────────────────────
# 3. CREAR GEODATAFRAME
# ─────────────────────────────────────────────────────────
gdf = gpd.GeoDataFrame(
    df,
    geometry=gpd.points_from_xy(df["x"], df["y"]),
    crs=CRS_UTM
)

# Cargar límite Huesca
huesca = gpd.read_file(GEOJSON_HUESCA).to_crs(CRS_UTM)
huesca_union = unary_union(huesca.geometry)

# Clip por si alguna coordenada se fue fuera
gdf = gdf[gdf.geometry.within(huesca_union)].copy()
print(f"  Granjas dentro del límite provincial: {len(gdf)}")

# Guardar GPKG
gdf.to_file(OUTPUT_GPKG, driver="GPKG", layer="granjas_porcino")
print(f"  GPKG guardado: {OUTPUT_GPKG}")

# ─────────────────────────────────────────────────────────
# 4. KDE ESPACIAL PONDERADO POR CAPACIDAD
# ─────────────────────────────────────────────────────────
print("Calculando KDE espacial ponderado por capacidad...")

cx = gdf["x"].values
cy = gdf["y"].values
w  = gdf["capacidad_total"].values.astype(float)
w  = w / w.sum()  # normalizar pesos

kde = gaussian_kde(
    np.vstack([cx, cy]),
    weights=w,
    bw_method="scott"   # ancho de banda automático — puedes probar 0.3, 0.5
)

# Grid de evaluación
b   = huesca_union.bounds  # minx, miny, maxx, maxy
xs  = np.arange(b[0], b[2], RES_M)
ys  = np.arange(b[1], b[3], RES_M)
XX, YY = np.meshgrid(xs, ys)
pts    = np.vstack([XX.ravel(), YY.ravel()])

print(f"  Evaluando {len(pts[0]):,} celdas ({RES_M}m resolución)...")
Z = kde(pts).reshape(XX.shape)

# ─────────────────────────────────────────────────────────
# 5. EXPORTAR RASTER GEOTIFF
# ─────────────────────────────────────────────────────────
tf = from_bounds(b[0], b[1], b[2], b[3], Z.shape[1], Z.shape[0])
with rasterio.open(
    OUTPUT_TIF, "w", driver="GTiff",
    height=Z.shape[0], width=Z.shape[1],
    count=1, dtype="float32",
    crs=CRS_UTM, transform=tf,
    compress="lzw"
) as dst:
    dst.write(Z.astype("float32"), 1)
print(f"  GeoTIFF guardado: {OUTPUT_TIF}")

# ─────────────────────────────────────────────────────────
# 6. MAPA PNG
# ─────────────────────────────────────────────────────────
print("Generando mapa...")

fig, ax = plt.subplots(figsize=(10, 12))

# Límite provincial
huesca.boundary.plot(ax=ax, color="black", linewidth=1)

# KDE como heatmap
extent = [b[0], b[2], b[1], b[3]]
im = ax.imshow(
    Z, extent=extent, origin="lower",
    cmap="YlOrRd", alpha=0.85,
    norm=mcolors.PowerNorm(gamma=0.4)   # realza zonas de baja densidad
)

# Puntos de granjas escalados por capacidad
scatter = ax.scatter(
    gdf["x"], gdf["y"],
    s=gdf["capacidad_total"] / 500,
    c="navy", alpha=0.4, linewidths=0,
    label="Granja (tamaño = capacidad)"
)

plt.colorbar(im, ax=ax, label="Densidad relativa (KDE)", shrink=0.6)
ax.set_title("Densidad porcina — Huesca\nFuente: REGA Aragón (CC BY 4.0)", fontsize=13)
ax.set_xlabel("X UTM (m)")
ax.set_ylabel("Y UTM (m)")
ax.legend(fontsize=9)
ax.set_aspect("equal")
plt.tight_layout()
plt.savefig(OUTPUT_PNG, dpi=150, bbox_inches="tight")
print(f"  PNG guardado: {OUTPUT_PNG}")

# ─────────────────────────────────────────────────────────
# RESUMEN
# ─────────────────────────────────────────────────────────
print()
print("=" * 50)
print("  RESUMEN FINAL")
print("=" * 50)
print(f"  Granjas procesadas:      {len(gdf):>8,}")
print(f"  Capacidad total:         {gdf['capacidad_total'].sum():>8,.0f} plazas")
print(f"  Dato real MAPA nov-2024: {5_099_794:>8,} cerdos")
print(f"  Ratio cap/real:          {gdf['capacidad_total'].sum()/5_099_794*100:>7.1f}%")
print(f"  Resolución KDE:          {RES_M} m")
print(f"  Bandwidth (Scott):       automático")
print("=" * 50)
print()
print("  Outputs:")
print(f"    {OUTPUT_TIF}  ← carga en QGIS")
print(f"    {OUTPUT_PNG}  ← vista rápida")
print(f"    {OUTPUT_GPKG} ← puntos con atributos")
```

```python
import numpy as np
import geopandas as gpd
import pandas as pd
import pydeck as pdk
from pyproj import Transformer
from shapely.geometry import mapping
import json
from pathlib import Path

MAP_DIR = Path("../../data/map/01_ganado_porcino")
MAP_DIR.mkdir(parents=True, exist_ok=True)

# --- 1. Límite provincia Huesca ---
print("Cargando límite provincia...")
DELIM_DIR = Path("../../data/raw/delimitations")
provincia = gpd.read_file(DELIM_DIR / "Huesca_Delimitacion.geojson").to_crs("EPSG:4326")

boundary_features = []
for geom in provincia.geometry:
    if geom.geom_type == "Polygon":
        boundary_features.append({
            "type": "Feature",
            "geometry": {"type": "Polygon", "coordinates": [list(geom.exterior.coords)]},
            "properties": {}
        })
    elif geom.geom_type == "MultiPolygon":
        for p in geom.geoms:
            boundary_features.append({
                "type": "Feature",
                "geometry": {"type": "Polygon", "coordinates": [list(p.exterior.coords)]},
                "properties": {}
            })

geojson_boundary = {"type": "FeatureCollection", "features": boundary_features}

layer_boundary = pdk.Layer(
    "GeoJsonLayer",
    data=geojson_boundary,
    stroked=True,
    filled=False,
    get_line_color=[255, 255, 255, 200],
    get_line_width=800,
    line_width_min_pixels=2,
    pickable=False,
)

# --- 2. Heatmap porcino ---
print("Preparando datos porcino...")
df_deck = df.dropna(subset=["x", "y", "capacidad_total"]).copy()

tr = Transformer.from_crs("EPSG:25830", "EPSG:4326", always_xy=True)
lon, lat = tr.transform(df_deck["x"].values, df_deck["y"].values)
df_deck["lon"]           = lon
df_deck["lat"]           = lat
df_deck["peso_sqrt"]     = np.sqrt(df_deck["capacidad_total"])
df_deck["capacidad_total"] = df_deck["capacidad_total"].astype(str)

heatmap_layer = pdk.Layer(
    "HeatmapLayer",
    data=df_deck,
    get_position=["lon", "lat"],
    get_weight="peso_sqrt",
    radius_pixels=40,
    intensity=2,
    threshold=0.03,
    opacity=0.9,
    color_range=[
        [  0,   0, 128,   0],
        [  0,   0, 255, 180],
        [  0, 255,   0, 200],
        [255, 165,   0, 220],
        [255,   0,   0, 240],
        [255, 255,   0, 255],
    ],
    pickable=False,
)

scatter_layer = pdk.Layer(
    "ScatterplotLayer",
    data=df_deck,
    get_position=["lon", "lat"],
    get_fill_color=[0, 0, 0, 0],
    get_radius=500,          # ← antes 3000, ahora mucho más pequeño
    radius_min_pixels=1,     # ← antes sin límite
    pickable=True,
)

# --- 3. Vista y deck ---
view = pdk.ViewState(
    longitude=-0.08,
    latitude=41.95,
    zoom=8,
    min_zoom=7,
    max_zoom=13,
    pitch=0,
    bearing=0,
)

tooltip = {
    "html": "<b>{explotacion}</b><br/>{municipio}<br/>Capacidad: {capacidad_total} plazas",
    "style": {"backgroundColor": "#111", "color": "white",
              "fontSize": "12px", "padding": "6px"},
}

deck = pdk.Deck(
    layers=[heatmap_layer, scatter_layer, layer_boundary],
    initial_view_state=view,
    tooltip=tooltip,
    map_style="https://basemaps.cartocdn.com/gl/dark-matter-gl-style/style.json",
)

leyenda_html = """
<div style="position:absolute; bottom:30px; left:20px; background:rgba(0,0,0,0.75);
            padding:14px 18px; border-radius:8px; font-family:sans-serif;
            font-size:13px; color:white; z-index:999; line-height:1.9;">
  <div style="font-weight:bold; font-size:14px; margin-bottom:8px;">🐷 Densidad porcina — Huesca</div>
  <div><span style="display:inline-block;width:14px;height:14px;background:#0000ff;border-radius:3px;margin-right:8px;vertical-align:middle;"></span>Baja</div>
  <div><span style="display:inline-block;width:14px;height:14px;background:#00ff00;border-radius:3px;margin-right:8px;vertical-align:middle;"></span>Media</div>
  <div><span style="display:inline-block;width:14px;height:14px;background:#ffa500;border-radius:3px;margin-right:8px;vertical-align:middle;"></span>Alta</div>
  <div><span style="display:inline-block;width:14px;height:14px;background:#ff0000;border-radius:3px;margin-right:8px;vertical-align:middle;"></span>Muy alta</div>
  <div><span style="display:inline-block;width:14px;height:14px;background:#ffff00;border-radius:3px;margin-right:8px;vertical-align:middle;"></span>Máxima</div>
</div>
"""

html_output = deck.to_html(as_string=True)
html_output = html_output.replace("</body>", leyenda_html + "</body>")

with open(MAP_DIR / "mapa_porcino_heatmap.html", "w", encoding="utf-8") as f:
    f.write(html_output)

print(f"Guardado: {MAP_DIR / 'mapa_porcino_heatmap.html'}")
```
